## Calculate Saturate Time point for each emotion

**Saturate time** are the time where the cumulative emotion counts reach 98% of their final sizes in our selected period dataset.

```python
df_2012_cum = df_2012.iloc[:,[0,7,8,9,10,11,12]]
df_2013_cum = df_2013.iloc[:,[0,7,8,9,10,11,12]]
```

```python
max_count_2013 = df_2013_cum.iloc[-1,:].tolist()[1:]
```

```python
max_count_2013
```

```python
def saturate_point(finish_count, df):
    saturate_count = finish_count * 0.98
    for i in reversed(range(df.shape[0])):
        if df.iloc[i,1] < saturate_count:
            return df.iloc[i+1,0]
```

```python
sp_sad_2013 = saturate_point(max_count_2013[0], df_2013_cum[['days_from_event', 'Sad_cum']])
sp_anx_2013 = saturate_point(max_count_2013[1], df_2013_cum[['days_from_event', 'Anx_cum']])
sp_ang_2013 = saturate_point(max_count_2013[2], df_2013_cum[['days_from_event', 'Anger_cum']])
sp_neg_2013 = saturate_point(max_count_2013[3], df_2013_cum[['days_from_event', 'Negemo_cum']])
sp_pos_2013 = saturate_point(max_count_2013[4], df_2013_cum[['days_from_event', 'Posemo_cum']])
sp_stress_2013 = saturate_point(max_count_2013[5], df_2013_cum[['days_from_event', 'Stress_cum']])
```

```python
print('Saturate Point for sadness 2013: ', sp_sad_2013)
print('Saturate Point for anxiety 2013: ', sp_anx_2013)
print('Saturate Point for anger 2013: ', sp_ang_2013)
print('Saturate Point for negative emotion 2013: ', sp_neg_2013)
print('Saturate Point for positive emotion 2013: ', sp_pos_2013)
print('Saturate Point for stress 2013: ', sp_stress_2013)
```

## Fit a smooth spline function to each emotion curve

We also draw a tangent line with the greatest slope in our fitted curve, and circle the time point where the growth curve of each emotion start deviating from its tangent line with the largest slope. We call this time point **Turning Point**. 

```python
from scipy.interpolate import UnivariateSpline
```

```python
def deriv(f,x):
    h = 0.000000001                 #step-size 
    return (f(x+h) - f(x))/h
```

```python
from matplotlib.patches import Circle

# spline for sadness 2013
x = df_2013_cum.iloc[:,0]
y = df_2013_cum.iloc[:,1]
f = UnivariateSpline(x, y, s=5000)

#x = np.linspace(-5,90,1000)
x_max = x.iloc[np.argmax(df_2013.loc[:,'Sad'])]
y_max = f(x_max)
y_tan = deriv(f,x_max) * (x - x_max) + y_max

plt.figure(figsize = (8,4))
plt.plot(x, f(x), '--')
plt.plot(x, y_tan,'b')
plt.plot(x, f(x), linewidth = 15, alpha = 0.2, color = 'gray')
plt.vlines(x = sp_sad_2013, ymin=0, ymax=max(f(x)), linestyles = 'dotted', color = 'red')
plt.text(x = 57, y = 2000, s = 'saturate at day 56', color='red')
plt.plot(3,1480,'bo',fillstyle='none',markersize=15)
plt.text(x=7, y=1460, s='turning point at day 3', color='blue')
plt.title('Spline for Sadness in 2013')
plt.xlabel('Days from event')
plt.ylabel('Cumulative tweets')
plt.ylim(0,max(y)*1.02)
```
